```python
from geopy.geocoders import Nominatim

def get_coordinates(city_name):
    geolocator = Nominatim(user_agent="my_geocoder")  # Replace with your user agent
    location = geolocator.geocode(city_name)
    
    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        return None, None

# Example usage
city_name = "New York"
latitude, longitude = get_coordinates(city_name)
if latitude is not None and longitude is not None:
    print(f"Latitude: {latitude}, Longitude: {longitude}")
else:
    print("Coordinates not found for the given city.")
```

```python
from tqdm import tqdm

db2['Lat'] = ""
db2['Long'] = ""

for index, row in tqdm(db2.iterrows(), total=len(db2)):
    loca = row['Headquarters Location'].split(',')
    latitude, longitude = get_coordinates(loca)
    db2.at[index, 'Lat'] = latitude
    db2.at[index, 'Long'] = longitude
```
